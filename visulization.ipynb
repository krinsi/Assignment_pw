#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
feature_1 = np.random.randn(100)
feature_2 = feature_1 + np.random.randn(100) * 0.5
feature_3 = feature_1 - feature_2 + np.random.randn(100) * 0.5

df = pd.DataFrame({
    'Feature_1': feature_1,
    'Feature_2': feature_2,
    'Feature_3': feature_3
})

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(20)

# Sample data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Aggregate the data
pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.1f', cbar_kws={'label': 'Average Sales'})
plt.title('Average Sales Heatmap by Month and Day')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the meshgrid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(df['X'].values.reshape(x.shape), df['Y'].values.reshape(y.shape), df['Z'].values.reshape(z.shape), cmap='viridis')

# Add color bar which maps values to colors
fig.colorbar(surf)

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.title('3D Surface Plot of $z = sin(sqrt(x^2 + y^2))$')
plt.show()
```

#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh to display plots in the notebook
output_notebook()

# Create x values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute y values as the sine of x
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
```

## 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Enable Bokeh to display plots in the notebook
output_notebook()

# Generate random data
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100

# Create random sizes and colors
sizes = np.random.randint(10, 40, size=n)
colors = np.random.randint(100, 255, size=n)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a new plot
p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer with sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=df, fill_alpha=0.6, line_color=None)

# Show the results
show(p)
```

# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Enable Bokeh to display plots in the notebook
output_notebook()

# Given dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot with a title and axis labels
p = figure(x_range=fruits,  title="Fruit Counts",toolbar_location=None, tools="")

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.8, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh to display plots in the notebook
output_notebook()

# Given data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot with title and axis labels
p = figure(title="Histogram of Data Distribution", background_fill_color="#fafafa")

# Plot the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Customize the plot
p.y_range.start = 0
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'
p.grid.grid_line_color="white"

# Show the plot
show(p)
```
