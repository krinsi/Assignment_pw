Question:1  Discuss string slicing and provide examples in python.

-Python slicing is about obtaining a sub-string from the given string by slicing it respectively from start to end.

-Python slicing can be done in two ways:
    1.Using a slice() method
    2.Using the array slicing  [:: ] method

1.Using a slice() method:
    
    -The slice() constructor creates a slice object representing the set of indices specified by range(start, stop, step).
    
    *Syntax:

        slice(stop)
        slice(start, stop, step)
        
    *Examples:
        
        String = 'PWSKILL'
        s1 = slice(3)
        s2 = slice(1, 5, 2)

        print(String[s1])
        print(String[s2])
        
2.Using the array slicing  [:: ] method:
    
    -In Python, indexing syntax can be used as a substitute for the slice object.
    -This is an easy and convenient way to slice a string using list slicing and Array slicing both syntax-wise and execution-wise.
    
    *Syntax:

        arr[start:stop]         
        arr[start:]             
        arr[:stop]              
        arr[:]
    
    *Examples:
        
        String = 'PWSKILL'
        print(String[:3])

Question-4: Compare and contrast tuples and lists with examples.

*Similarities:

    1.Ordered Collections:  Both tuples and lists are ordered collections of items.
                            The order in which items are inserted is maintained.

    2.Indexing and Slicing: Both support indexing and slicing to access elements.

    3.Heterogeneous:        Both can store items of different data types.
    
*Differences:

    1.Mutability:

        List: Mutable, meaning you can change, add, or remove items after the list has been created.
        Tuple: Immutable, meaning once a tuple is created, you cannot change, add, or remove items.
        
    2.Syntax:

        List: Defined using square brackets [].
        Tuple: Defined using parentheses ().
        
*List Examples:
    
    fruits = ['apple', 'banana', 'cherry']
    print(fruits[1])
    fruits[1] = 'blueberry'
    print(fruits)

    fruits.remove('apple')
    print(fruits)
    
*Tuple Example:
    
    fruits = ('apple', 'banana', 'cherry')

Question-5: Describe the key features of sets and provide examples of their use.

*Key Features of Sets:
    
1.Unordered Collection:
        Sets do not maintain the order of elements. The items have no index.
2.Unique Elements:
        All elements in a set must be unique. Duplicate elements are automatically removed.
3.Mutable:
        Sets are mutable, meaning you can add and remove elements after the set has been created.
4.No Indexing:
        Since sets are unordered, they do not support indexing or slicing.
        
*Creating a Set:
    -You can create a set using curly braces {} or the set() function.
    
    *Example:
            fruits = {'apple', 'banana', 'cherry'}
            print(fruits)
            
*Adding Elements to a Set:
    -You can add elements to a set using the add() method.
  
    *Example:
            fruits = {'apple', 'banana', 'cherry'}
            fruits.add('date')
            print(fruits)
            
*Removing Elements from a Set:
    -You can remove elements from a set using the remove() or discard() methods.
    -The remove() method will raise an error if the element is not found, whereas discard() will not.
    
    *Example:
            fruits = {'apple', 'banana', 'cherry'}
            fruits.remove('banana')
            print(fruits)
            
****Set Operations:******

1.Union:
    -The union of two sets is a set containing all elements from both sets.
    *Example:
        set1 = {'apple', 'banana'}
        set2 = {'banana', 'cherry'}
        union_set = set1.union(set2)
        print(union_set)
        
2.Intersection:
    -The intersection of two sets is a set containing only the elements that are present in both sets.
    *Example:
        set1 = {'apple', 'banana'}
        set2 = {'banana', 'cherry'}
        intersection_set = set1.intersection(set2)
        print(intersection_set)
        
3.Difference:
    -The difference of two sets is a set containing elements that are in the first set but not in the second set.
    *Example:
        set1 = {'apple', 'banana'}
        set2 = {'banana', 'cherry'}
        difference_set = set1.difference(set2)
        print(difference_set)

Question-7: Describe how to add,modify and delete items in a dictionary with examples.

-In Python, dictionaries are mutable data structures that store key-value pairs.
-They allow you to add, modify, and delete items easily.

1.Adding Items to a Dictionary:
    
    -To add an item to a dictionary, you can simply assign a value to a new key.
    *Example:
        person = {'name': 'John', 'age': 30}
        person['city'] = 'New York'
        
2.Modifying Items in a Dictionary:

        -To modify an item in a dictionary, you can assign a new value to an existing key.
        *Example:
            person['age'] = 31
            print(person)
            
3.Deleting Items from a Dictionary:
    
    -There are several methods to delete items from a dictionary:

    3.1 Using del statement:
        del person['city']
        print(person)  # Output: {'name': 'John', 'age': 31}

    3.2 Using pop() method:

    -The pop() method removes the item with the specified key and returns its value.
            age = person.pop('age')
            print(age)
            print(person)
    3.3 Using popitem() method:

    -The popitem() method removes and returns the last inserted key-value pair as a tuple.
        person['city'] = 'New York'
        item = person.popitem()
        print(item)
        print(person)
        
    3.4 Using clear() method:

    -The clear() method removes all items from the dictionary.
        person['age'] = 31
        person['city'] = 'New York'
        person.clear()
        print(person)

Question-8: Discuss the importance of dictionary keys being Immttable and provide examples.

********Importance of Immutable Dictionary Keys******

1.Consistency and Reliability:

    -If dictionary keys were mutable, their values could change, leading to inconsistencies and making it difficult to locate
      the keys associated value.
    -Immutable keys guarantee that once a key is placed in a dictionary, it remains unchanged.
    
2.Hashability:

    -Dictionaries use a hash table to store and retrieve values efficiently.
    -Immutable objects like strings, numbers, and tuples have a fixed hash value, making them suitable for use as dictionary keys.
    -Mutable objects, like lists and dictionaries, do not have a fixed hash value because their content can change, leading to potential
     errors and inefficiencies in the hash table.
        
3.Performance:

    -Using immutable keys ensures that the dictionarys performance remains optimal.
    -The hash function can quickly locate the keys position, allowing for O(1) average time complexity for lookups, inserts, and deletions.
    
******Examples******
-Using Immutable Keys

1.Strings as Keys:
    
    phone_book = {'Alice': '123-456-7890', 'Bob': '987-654-3210'}
    print(phone_book['Alice'])
    
2.Numbers as Keys:

    square_numbers = {1: 1, 2: 4, 3: 9, 4: 16}
    print(square_numbers[3])  

3.Tuples as Keys:

    coordinates = {(0, 0): 'Origin', (1, 2): 'Point A', (3, 4): 'Point B'}
    print(coordinates[(1, 2)])
